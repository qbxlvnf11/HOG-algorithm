```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('copyright-violated-deleted-royalty-free-thumbnail.jpg')
image = np.float32(image) / 255.0

# Converting BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Converting RGB to Gray scale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
plt.figure(figsize=(15, 8))
plt.title("Original Gray Image " + str(image.shape))
plt.imshow(gray_image, cmap="gray")
plt.axis("off")
plt.show()
```

#### - Calculation gradient

```python
gx = cv2.Sobel(gray_image, cv2.CV_32F, 1, 0, ksize=1)
gy = cv2.Sobel(gray_image, cv2.CV_32F, 0, 1, ksize=1)
```

#### - Calculation gradient magnitude and orientation

```python
magnitude, angle = cv2.cartToPolar(gx, gy, angleInDegrees=True)
```

```python
print('magnitude:', magnitude)
plt.figure(figsize=(15, 8))
plt.imshow(magnitude, cmap="gray")
plt.axis("off")
plt.show()
```

```python
print('angle:', angle)
plt.figure(figsize=(15, 8))
plt.imshow(angle, cmap="gray")
plt.axis("off")
plt.show()
```
